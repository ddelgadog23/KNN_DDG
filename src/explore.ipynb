```python
df = df.drop('fixed acidity', axis= 1)
df.info()
```

```python
df.describe()
```

```python
# Sintaxis: df.loc[filas, columnas]
valores_altos = df.loc[df['citric acid'] > 0.8, 'citric acid']
print(valores_altos)
```

```python
df = df[df['citric acid']<0.8]
```

```python
df = df[df['sulphates']<1.5]
```

```python
# Escalado de Valores

from sklearn.preprocessing import MinMaxScaler
x = df.drop('objetivo',axis=1)

scaler = MinMaxScaler()

X_scaled = pd.DataFrame(scaler.fit_transform(x), index = df.index, columns=x.columns)
X_scaled.head()
```

```python
from sklearn.model_selection import train_test_split

x = X_scaled
y = df['objetivo']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
model.fit(x_train,y_train)

```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(x_test)

print(f'La precision de modelo es: {accuracy_score(y_test,y_pred)}')
```

```python
# Obtencion del numero ideal de K
accuracies = []
for k in range(1,21):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    accuracies.append(accuracy_score(y_test,y_pred))

plt.plot(accuracies)
plt.tight_layout()
plt.show()
    
```

```python
model_k1 = KNeighborsClassifier(n_neighbors=3)
model_k1.fit(x_train,y_train)
y_pred2 = model_k1.predict(x_test)
accuracy_score(y_test,y_pred2)
```

- Segun el bucle el mejor valor k para determinar el numeor de vecinos en 2.5, al llevarlo a 3 aumento la precision del modelo por default que utilizo el valor de 5.
